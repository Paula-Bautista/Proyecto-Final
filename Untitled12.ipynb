```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import pandas as pd
```

La primera gráfica muestra cómo transforma una entrada la función de activación sigmoide en una neurona artificial.

Ecuación usada:
sigmoid(x) = 1 / (1 + exp(-x))

```python
x = np.linspace(-10, 10, 500)
sigmoid = 1 / (1 + np.exp(-x))

plt.figure(figsize=(7,5))
plt.plot(x, sigmoid)
plt.title("Función de Activación Sigmoide")
plt.xlabel("x")
plt.ylabel("sigmoid(x)")
plt.grid(True)
plt.show()
```

La línea es la frontera de decisión del perceptrón.
Divide el plano en 2 clases.

Ecuación usada:
w1x1 + w2x2 + b = 0
Acá usamos: 1x1 – 1x2 + 0 = 0 → x2 = x1

Qué significa:
Todo punto sobre la línea es indeciso.

Por encima → clase 1

Por debajo → clase 0

Representa cómo una neurona simple separa datos en un plano.

```python
w1, w2, b = 1, -1, 0
x_vals = np.linspace(-5, 5, 200)
y_vals = (w1*x_vals + b) / (-w2)

plt.figure(figsize=(7,5))
plt.plot(x_vals, y_vals)
plt.title("Frontera de Decisión de un Perceptrón")
plt.xlabel("x1")
plt.ylabel("x2")
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.grid(True)
plt.show()
```

Es una red neuronal con:

2 entradas

2 neuronas ocultas

1 salida

Ecuaciones implicadas:
Esta gráfica muestra la forma general de la operación:

h_j = activación(w*x + b)

y = activación(w*h + b)

Qué significa:
Representa cómo fluye la información:
entrada → capa oculta → salida.

La dirección de cada flecha es el recorrido de la señal.

```python
G = nx.DiGraph()

G.add_nodes_from([
    ("x1", {"layer": "input"}),
    ("x2", {"layer": "input"}),
    ("h1", {"layer": "hidden"}),
    ("h2", {"layer": "hidden"}),
    ("y", {"layer": "output"})
])

edges = [
    ("x1", "h1"), ("x1", "h2"),
    ("x2", "h1"), ("x2", "h2"),
    ("h1", "y"),  ("h2", "y")
]

G.add_edges_from(edges)

pos = {
    "x1": (-1, 0),
    "x2": (-1, -1),
    "h1": (0, -0.3),
    "h2": (0, -0.8),
    "y":  (1, -0.55)
}

plt.figure(figsize=(8,6))
nx.draw(G, pos, with_labels=True, node_size=2000)
plt.title("Pequeña Red Neuronal: Propagación de Señal")
plt.show()
```

Qué representa:
El descenso del gradiente intenta encontrar el mínimo de la función.
Cost = theta^2

Ecuación usada:
J(theta) = theta²
dJ/dtheta = 2*theta

Qué significa:

La curva sube hacia ambos lados.

El mínimo está en θ = 0.

Muestra cómo el entrenamiento busca minimizar errores moviéndose “hacia abajo”.

```python
x = np.linspace(-3, 3, 400)
cost = x**2

plt.figure(figsize=(7,5))
plt.plot(x, cost)
plt.title("Función de Costo J(θ)")
plt.xlabel("θ")
plt.ylabel("J(θ)")
plt.grid(True)
plt.show()
```

Qué representa:
Simula cómo disminuye el error mientras la red aprende.

Ecuación usada:
error(época) = exp(-época/10) + ruido_pequeño

Qué significa:

Al principio el error es alto.

Luego baja rápido.

Finalmente se estabiliza.

Esto es EXACTAMENTE lo que se espera ver en un entrenamiento real.

```python
epochs = np.arange(1, 51)
error = np.exp(-epochs/10) + 0.05*np.random.rand(50)

plt.figure(figsize=(7,5))
plt.plot(epochs, error)
plt.title("Disminución del Error Durante el Entrenamiento")
plt.xlabel("Época")
plt.ylabel("Error")
plt.grid(True)
plt.show()
```
