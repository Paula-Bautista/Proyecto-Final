```python
# archivo: redes_mielina_sim.py
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

# reproducibilidad
np.random.seed(42)

# ---------- UTILIDADES ----------
def rk4_step_scalar(f, y, t, dt):
    k1 = f(y, t)
    k2 = f(y + 0.5*dt*k1, t + 0.5*dt)
    k3 = f(y + 0.5*dt*k2, t + 0.5*dt)
    k4 = f(y + dt*k3, t + dt)
    return y + (dt/6.0)*(k1 + 2*k2 + 2*k3 + k4)

# ---------- CONECTIVIDAD ----------
def build_nx_connectivity(Nx, Ny, p):
    """Crea un grafo dirigido sobre la malla 2D con prob p por arista (ambas direcciones con prob p)."""
    Ggrid = nx.grid_2d_graph(Nx, Ny)  # grafo no dirigido de la cuadrícula
    H = nx.DiGraph()
    H.add_nodes_from(Ggrid.nodes())
    for u, v in Ggrid.edges():
        if np.random.rand() < p:
            H.add_edge(u, v)
        if np.random.rand() < p:
            H.add_edge(v, u)
    return H

def build_mask_networkx(N, M, p):
    G = build_nx_connectivity(N, M, p)
    mask = {}
    for i in range(N):
        for j in range(M):
            mask[(i, j)] = list(G.successors((i, j)))
    return mask

# ---------- LIF NETWORK (discreto, con acoplamiento sináptico sencillo) ----------
def lif_rhs_scalar(V, t, Cm, gL, EL, Iext, Isyn):
    return (-gL*(V - EL) + Iext + Isyn)/Cm

def simulate_lif_network(Nx, Ny, T, dt, params, G, mu_t, stim):
    steps = int(np.round(T/dt))
    V = np.full((Nx, Ny), params['Vrest'], dtype=float)
    Vtrace = np.zeros((steps, Nx, Ny), dtype=float)
    fired_count = np.zeros((Nx, Ny), dtype=int)
    refr = np.zeros((Nx, Ny), dtype=float)

    # precompute neighbor lists for velocidad
    succ = {node: list(G.successors(node)) for node in G.nodes()}

    for step in range(steps):
        t = step*dt
        Isyn = np.zeros((Nx, Ny), dtype=float)
        spikes = (V >= params['Vth']).astype(float)

        # actualizar conteo de disparos y periodo refractario y reset
        fired_count += spikes.astype(int)
        if np.any(spikes):
            V[spikes > 0] = params['Vreset']
            refr[spikes > 0] = params['tref']

        # propagar sinápticamente (uso mu_t[step] como factor global de eficacia)
        w = params.get('w', 1.0)
        efficacy = mu_t[step] if step < len(mu_t) else mu_t[-1]
        # iterar solo sobre los que dispararon
        fired_idx = np.argwhere(spikes > 0)
        for (i, j) in fired_idx:
            src = (i, j)
            for dst in succ.get(src, []):
                a, b = dst
                Isyn[a, b] += efficacy * w

        # estímulo externo para este paso: admite matriz o escalar
        Iext_step = stim[step] if stim.shape[0] > step else np.zeros((Nx, Ny))

        # integración por RK4 (por neurona, escalar)
        for i in range(Nx):
            for j in range(Ny):
                if refr[i, j] > 0:
                    refr[i, j] -= dt
                    continue
                f = lambda v, tt: lif_rhs_scalar(v, tt, params['Cm'], params['gL'], params['EL'],
                                                 Iext_step[i, j], Isyn[i, j])
                V[i, j] = rk4_step_scalar(f, V[i, j], t, dt)

        Vtrace[step] = V

    return Vtrace, fired_count

# ---------- FHN 1D (modelo excitable con acoplamiento difusivo) ----------
def simulate_fhn_1d(Nx, T, dt, params, mu_profile, stimulus):
    """
    Simula FitzHugh-Nagumo en 1D con acoplamiento difusivo lineal:
    dV/dt = (V - V^3/3 - W + Isyn)/eps
    dW/dt = V + a - b*W
    Isyn_i = g * mu_profile[i] * (V_{i-1} - V_i) + g * mu_profile[i] * (V_{i+1} - V_i)
    Condiciones de frontera: Neumann (no-flux)
    """
    steps = int(np.round(T/dt))
    V = np.full(Nx, params['V0'], dtype=float)
    W = np.full(Nx, params['W0'], dtype=float)
    M = np.zeros((steps, Nx), dtype=float)

    eps = params['eps']
    a = params['a']
    b = params['b']
    g = params.get('g', 0.0)

    for step in range(steps):
        t = step*dt
        # compute diffusion-like Isyn from neighbors (vectorizado)
        V_left = np.roll(V, 1)
        V_right = np.roll(V, -1)
        # No-flux: reflect boundaries
        V_left[0] = V[0]
        V_right[-1] = V[-1]
        # mu_profile may be scalar or array
        mu = mu_profile if np.ndim(mu_profile) == 1 else np.full(Nx, mu_profile)
        Isyn = g * mu * ((V_left - V) + (V_right - V))

        # stimulus can be per-space array at this time step
        Iext = stimulus[step] if (hasattr(stimulus, 'shape') and stimulus.shape[0] > step) else np.zeros(Nx)

        # RHS functions for RK4 (vectorized per node but using scalar RK4)
        def rhs_scalar(idx):
            def rhs_v(v, tt):
                # temporarily construct local V_i and W_i; here we treat coupling as additive constant Isyn[idx]
                return (v - v**3/3.0 - W[idx] + Isyn[idx] + Iext[idx]) / eps
            return rhs_v

        # Integrate V with RK4 scalar-wise; W with Euler (stiffness low for chosen dt)
        V_new = V.copy()
        W_new = W.copy()
        for i in range(Nx):
            V_new[i] = rk4_step_scalar(rhs_scalar(i), V[i], t, dt)
            # integrate W (simple Euler)
            W_new[i] = W[i] + dt*(V[i] + a - b*W[i])

        V, W = V_new, W_new
        M[step] = V

    return M

# ---------- ECUACIÓN DE CABLE 1D (discretización explícita) ----------
def cable_1d(Nx, dx, T, dt, Cm, Ri, Rm, V0, Iext_time_space):
    steps = int(np.round(T/dt))
    V = np.full(Nx, V0, dtype=float)
    M = np.zeros((steps, Nx), dtype=float)
    # discretización explícita: dV/dt = (1/(Ri*Cm)) * d2V/dx2 - V/(Rm*Cm) + Iext/Cm
    coef_diff = dt/(Ri*Cm*dx*dx)
    coef_leak = dt/(Rm*Cm)
    for s in range(steps):
        lap = np.zeros_like(V)
        lap[1:-1] = V[2:] - 2*V[1:-1] + V[:-2]
        # condiciones de frontera Neumann (no flux): lap[0]=V[1]-V[0], lap[-1]=V[-2]-V[-1]
        lap[0] = V[1] - V[0]
        lap[-1] = V[-2] - V[-1]
        Iext = Iext_time_space[s] if (hasattr(Iext_time_space, 'shape') and Iext_time_space.shape[0] > s) else np.zeros(Nx)
        V = V + coef_diff*lap - coef_leak*V + (dt/Cm)*Iext
        M[s] = V
    return M

# ---------- ARENA DE ARENA (sandpile) ----------
def sandpile_step(E, Ec, mask, mu):
    N, M = E.shape
    topple = (E > Ec)
    if not topple.any():
        return E, 0
    E2 = E.copy()
    total = 0
    coords = np.argwhere(topple)
    for (i, j) in coords:
        nbrs = mask.get((i, j), [])
        k = len(nbrs)
        if k > 0:
            share = mu * E[i, j] / k
            for (a, b) in nbrs:
                E2[a, b] += share
        E2[i, j] = 0.0
        total += 1
    return E2, total

def run_sandpile(E0, T, Ec, mask, mu_series, drive):
    E = E0.copy().astype(float)
    N, M = E.shape
    aval = []
    act = []
    for s in range(T):
        a = np.random.randint(0, N)
        b = np.random.randint(0, M)
        E[a, b] += drive
        size = 0
        while True:
            E, t = sandpile_step(E, Ec, mask, mu_series[s] if s < len(mu_series) else mu_series[-1])
            if t == 0:
                break
            size += t
        aval.append(size)
        act.append(np.mean(E > 0))
    return np.array(aval), np.array(act)

# ---------- VELOCIDAD ESTIMADA ----------
def estimate_velocity(Vmap, dx, dt):
    # Vmap: (time, space)
    steps, Nx = Vmap.shape
    # instante de pico por posición (siempre en [0, steps-1])
    pk = np.argmax(Vmap, axis=0)
    x = np.arange(Nx) * dx
    t = pk * dt
    valid = (t > 0)
    if np.sum(valid) < 2:
        return np.nan
    c = np.polyfit(x[valid], t[valid], 1)  # t = slope * x + intercept
    slope = c[0]
    if slope == 0:
        return np.nan
    return 1.0 / slope

# ---------- PARÁMETROS Y EJECUCIONES (gráficos inmediatamente después) ----------
if __name__ == '__main__':
    # ---------- LIF NETWORK ----------
    Nx, Ny = 40, 40
    T = 200.0
    dt = 0.05
    steps = int(np.round(T/dt))

    params_lif = {'Cm': 1.0, 'gL': 0.1, 'EL': -65.0, 'Vrest': -65.0,
                  'Vth': -50.0, 'Vreset': -65.0, 'tref': 2.0, 'w': 4.0}
    G = build_nx_connectivity(Nx, Ny, p=0.5)

    mu_t = np.ones(steps)  # eficacia sináptica en tiempo
    stim = np.zeros((steps, Nx, Ny))
    cx, cy = Nx // 2, Ny // 2
    # pulso breve en el centro
    for s in range(steps):
        if int(20/dt) <= s < int(23/dt):
            stim[s, cx, cy] = 30.0

    Vtrace, fired = simulate_lif_network(Nx, Ny, T, dt, params_lif, G, mu_t, stim)

    # gráfica inmediatamente después
    tvec = np.linspace(0, T, steps)
    plt.figure(figsize=(7, 4))
    plt.plot(tvec, Vtrace[:, cx, cy])
    plt.xlabel("Tiempo (ms)")
    plt.ylabel("Voltaje (mV)")
    plt.title("Potencial temporal - neurona central (LIF)")
    plt.tight_layout()
    plt.savefig("fig1_lif.png", dpi=200)
    plt.show()

    # ---------- FHN 1D ----------
    Nx1 = 200
    T1 = 120.0
    dt1 = 0.02
    steps1 = int(np.round(T1/dt1))

    params_fhn = {'eps': 0.08, 'a': 0.7, 'b': 0.8, 'V0': -1.0, 'W0': 1.0, 'g': 0.1}
    mu_prof = np.ones(Nx1) * 0.9
    stim_profile = np.zeros((steps1, Nx1))
    for s in range(steps1):
        if int(4/dt1) <= s < int(4.4/dt1):
            stim_profile[s, 5] = 1.2

    Vmap_fhn = simulate_fhn_1d(Nx1, T1, dt1, params_fhn, mu_prof, stim_profile)

    plt.figure(figsize=(8, 4))
    plt.imshow(Vmap_fhn, aspect='auto', cmap='inferno', origin='lower',
               extent=[0, Nx1-1, 0, T1])
    plt.colorbar(label='V')
    plt.xlabel("Posición")
    plt.ylabel("Tiempo (ms)")
    plt.title("Mapa FHN (1D)")
    plt.tight_layout()
    plt.savefig("fig2_fhn.png", dpi=200)
    plt.show()

    # ---------- CABLE 1D ----------
    Nx_c = 300
    dx = 0.1
    dtc = 0.01
    T_c = 40.0
    steps_c = int(np.round(T_c/dtc))
    Iext = np.zeros((steps_c, Nx_c))
    for s in range(steps_c):
        if int(2/dtc) <= s < int(2.4/dtc):
            Iext[s, 10] = 40.0

    Vmap_c = cable_1d(Nx_c, dx, T_c, dtc, Cm=1.0, Ri=80.0, Rm=20000.0, V0=-65.0, Iext_time_space=Iext)

    plt.figure(figsize=(8, 4))
    plt.imshow(Vmap_c, aspect='auto', origin='lower', cmap='viridis', extent=[0, Nx_c-1, 0, T_c])
    plt.colorbar(label='V (mV)')
    plt.title("Cable equation 1D")
    plt.xlabel("Posición")
    plt.ylabel("Tiempo (ms)")
    plt.tight_layout()
    plt.savefig("fig3_cable.png", dpi=200)
    plt.show()

    # ---------- VELOCIDAD vs MIELINA (usando FHN reducido) ----------
    mus = np.linspace(0.0, 1.0, 18)
    vels = []
    # usar un dominio más corto para barrido rápido
    Nx_v = 120
    T_v = 70.0
    dt_v = 0.05
    for m in mus:
        mu_prof2 = np.ones(Nx_v) * m
        stim_tmp = np.zeros((int(np.round(T_v/dt_v)), Nx_v))
        # breve pulso en la posición 5
        for s in range(stim_tmp.shape[0]):
            if int(4/dt_v) <= s < int(4.4/dt_v):
                stim_tmp[s, 5] = 1.2
        Vm = simulate_fhn_1d(Nx_v, T_v, dt_v, params_fhn, mu_prof2, stim_tmp)
        vel = estimate_velocity(Vm, dx=1.0, dt=dt_v)
        vels.append(vel)

    plt.figure(figsize=(7, 4))
    plt.plot(mus, vels, marker='o')
    plt.xlabel("μ (eficacia / mielina simulada)")
    plt.ylabel("Velocidad estimada")
    plt.title("Velocidad vs mielina (FHN 1D)")
    plt.tight_layout()
    plt.savefig("fig4_vel_vs_m.png", dpi=200)
    plt.show()

    # ---------- SANDPILE (avalanchas) ----------
    Nsp, Msp = 70, 70
    E0 = np.zeros((Nsp, Msp))
    mask = build_mask_networkx(Nsp, Msp, p=0.5)
    Tsp = 2000  # reducir tiempo para ejecución rápida
    mu_series = np.ones(Tsp) * 0.5
    aval_c, act_c = run_sandpile(E0, Tsp, Ec=1.0, mask=mask, mu_series=mu_series, drive=0.4)

    mu_series2 = np.maximum(0.0, 0.5 - 0.5*(np.arange(Tsp)/Tsp)**1.4)
    aval_d, act_d = run_sandpile(E0, Tsp, Ec=1.0, mask=mask, mu_series=mu_series2, drive=0.4)

    plt.figure(figsize=(7, 4))
    bins = np.logspace(0, np.log10(max(np.max(aval_c), np.max(aval_d)) + 1), 50)
    plt.hist(aval_c[aval_c>0], bins=bins, alpha=0.6, label='const mu')
    plt.hist(aval_d[aval_d>0], bins=bins, alpha=0.6, label='decay mu')
    plt.xscale("log")
    plt.yscale("log")
    plt.title("Avalanchas (sandpile)")
    plt.legend()
    plt.tight_layout()
    plt.savefig("fig5_avalanches.png", dpi=200)
    plt.show()
```
