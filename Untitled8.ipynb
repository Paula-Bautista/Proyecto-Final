```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
h = 0.01
x0 = 0.0
y0 = 2.0
x_max = 5.0
n = int((x_max - x0)/h)
```

```python
def f(x, y):
  return x - 2*y
```

```python
x_vals = np.zeros(n+1)
y_vals = np.zeros(n+1)
x_vals[0] = x0
y_vals[0] = y0
for i in range(n):
    x_i = x_vals[i]
    y_i = y_vals[i]

    k1 = f(x_i, y_i)
    k2 = f(x_i + h/3, y_i + h/3*k1)
    k3 = f(x_i + h/3, y_i + h/6*(k1 + k2))
    k4 = f(x_i + h/2, y_i + h/8*(k1 + 3*k3))
    k5 = f(x_i + h, y_i + h/2*(-3*k1 + 2*k2 + 12*k3 - 12*k4))
    k6 = f(x_i + h, y_i + h/7*(7*k1 + 10*k3 + k4 + 2*k5))

    y_vals[i+1] = y_i + h/90*(7*k1 + 32*k3 + 12*k4 + 32*k5 + 7*k6)
    x_vals[i+1] = x_i + h
```

```python
y_exact = 0.25*(2*x_vals + 9*np.exp(-2*x_vals) - 1)
```

```python
plt.figure(figsize=(8,5))
plt.plot(x_vals, y_vals, color='#8000FF', label='Método de Butcher')
plt.plot(x_vals, y_exact, color='#00B050', linestyle='--', label='Solución analítica')
plt.title('Comparación: Método de Butcher vs Solución Analítica')
plt.xlabel('x')
plt.ylabel('y(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
m = 1.0
k = 1.0
gamma0 = 0.1
gamma1 = 0.05
F0 = 2.0
omega = 1.25
h = 0.001
t_max = 100
t = np.arange(0, t_max, h)
x0 = 0.0
v0 = 0.0
a0 = 0.0

def f(t):
    return F0 * np.sin(omega * t)
```

```python
def sistema(t, y):
    y1, y2, y3 = y
    dy1 = y2
    dy2 = y3
    dy3 = (f(t) - m * y3 - gamma0 * y2 - k * y1) / gamma1
    return np.array([dy1, dy2, dy3])
```

```python
def butcher_step(t, y, h):
    k1 = sistema(t, y)
    k2 = sistema(t + h/3, y + h/3*k1)
    k3 = sistema(t + h/3, y + h/6*(k1 + k2))
    k4 = sistema(t + h/2, y + h/8*(k1 + 3*k3))
    k5 = sistema(t + h, y + h/2*(-3*k1 + 2*k2 + 12*k3 - 12*k4))
    k6 = sistema(t + h, y + h/7*(7*k1 + 10*k3 + k4 + 2*k5))
    return y + h/90*(7*k1 + 32*k3 + 12*k4 + 32*k5 + 7*k6)
Y = np.zeros((len(t), 3))
Y[0] = [x0, v0, a0]

for i in range(len(t)-1):
    Y[i+1] = butcher_step(t[i], Y[i], h)

x = Y[:, 0]
v = Y[:, 1]
a = Y[:, 2]
```

```python
plt.figure(figsize=(10,5))
plt.plot(t, x, color='orange', label="x(t)")
plt.plot(t, v, color='green', label="x'(t)")
plt.plot(t, a, color='purple', label="x''(t)")
plt.xlabel("Tiempo (t)")
plt.ylabel("Parametro")
plt.title('Oscilador con memoria efectiva)')
plt.legend()
plt.grid(True)
plt.show()
```
